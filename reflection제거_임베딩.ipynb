```python
import pandas as pd
import numpy as np
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer('jhgan/ko-sroberta-multitask')

sentences = ["안녕하세요?", "한국어 문장 임베딩을 위한 버트 모델입니다."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
data = pd.read_csv('/content/drive/Shareddrives/DA 시니어 3차프로젝트/data/total_without_reflection.csv')
data
```

```python
data = data[-data['reflection_and_response'].isna()]
data
```

```python
data.loc[0, 'sentence']
```

```python
model.encode(data.loc[0, 'sentence'])
```

```python
%%time
data['embedding'] = data['sentence'].map(lambda x : list(model.encode(x)))
```

```python
data
```

```python
data.to_csv('/content/drive/Shareddrives/DA 시니어 3차프로젝트/data/total_embeded_without_reflection.csv', index = False)
```

```python
text = input('대화 입력 : ')
```

```python
embedding = model.encode(text)
```

```python
data['distance'] = data['embedding'].map(lambda x : cosine_similarity([embedding], [x]).squeeze())
```

```python
data.head()
```

```python
answer = data.loc[data['distance'].idxmax()]
```

```python
answer
```

```python
print('구분 :', answer['Total category'])
print('가장 유사한 유저질문 :', answer['sentence'])
print('챗봇 답변 :', answer['reflection_and_response'])
print('코사인유사도 :', answer['distance'])
```
