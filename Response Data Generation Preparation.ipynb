```python
from pororo import Pororo
mt = Pororo(task="translation", lang="multi")
```

```python
def paraphrase(text) :
    text_to_eng = mt(text, src="ko", tgt="en")
    text_to_kor = mt(text_to_eng, src="en", tgt="ko")
    return text_to_kor
```

## Define 종결어미 변환 function

```python
pip install konlpy
```

```python
import pandas as pd
import numpy as np
import random
import konlpy
import re
from konlpy.tag import Komoran
komoran = Komoran()
```

```python
sample_sentence_1 = "나도 이러기 싫은데 내 마음대로 안돼."
sample_sentence_2 = "날 부르고 있다." 
sample_sentence_3 = "순간순간 감정조절을 못 한다."
```

```python
def change_eomi (sentence):
  string = re.findall(r'\w+', sentence)
  word = string[-1]
  morphtags = Komoran().pos(word)

  if morphtags[0][1] == 'VA' or morphtags[0][1] == 'VV' or morphtags[0][1] == 'VXV' or morphtags[0][1] == 'VXA' or morphtags[0][1] == 'VCP' or morphtags[0][1] == 'VCN': #lemmatize
    changed_word = str(morphtags[0][0] + '군요.')
    changed_sentence = ' '.join(string[:-1]) + ' ' + changed_word
    return changed_sentence

  else :
      candidate = ['그런 일이 있으셨군요.', '그랬군요.', '그러셨군요.']
      sentence = np.random.choice(candidate)
      return sentence
```

```python
print(change_eomi(sample_sentence_1))
print(change_eomi(sample_sentence_2))
print(change_eomi(sample_sentence_3))
```

## Define Reflection response function

```python
def reflect(text):
    paraphrased_text = paraphrase(text)
    final_text = change_eomi(paraphrased_text)
    return final_text
```

```python
sample_sentence_1 = "나도 이러기 싫은데 내 마음대로 안돼."
sample_sentence_2 = "근데 한편으론 내가 왜 이러나 걱정이 되기도 해요." 
sample_sentence_3 = "요즘 이거 때문에 잠도 안 오고 진짜 걱정돼서 미치겠다."
```

```python
print(reflect(sample_sentence_1))
print(reflect(sample_sentence_2))
print(reflect(sample_sentence_3))
```

## Change Pronouns

```python
ex1 = "나는 더 이상 감정을 통제할 수 없어요."
ex2 = "내 감정을 나도 모르겠어요."
ex3 = "막 내 감정을 내가 주체 못 한다나? 화 내는 게 뭔가 평소랑 다르대."
ex4 = "나도 이러고 싶지 않은데 내 맘대로 안 돼."
ex5 = "제 감정이 이상해진 것 같아요. 남편만 보면 화가 치밀어 오르고 감정 조절이 안돼요."
```

```python
change_pronoun = pd.DataFrame()
change_pronoun['Na_list'] = ['나는', '나도', '내', '나의', '내가']
change_pronoun['DangShin_list'] = ['당신은', '당신도', '당신의', '당신의', '당신이']
```

```python
change_pronoun
```

```python
string = re.findall(r'\w+', ex2)
pronoun_changed_sentence = []

for tokens in string:
    if tokens in ['나는', '나도', '내', '나의', '내가']:
        get_index = change_pronoun['Na_list'][change_pronoun['Na_list'] == tokens].index
        index = int(*get_index)
        tokens = change_pronoun['DangShin_list'][index]
    pronoun_changed_sentence.append(tokens)

print(*pronoun_changed_sentence)
```

```python
def change_eomi_and_pronoun (sentence):
    string = re.findall(r'\w+', sentence)

    if string != [] :
        #pronoun change
        subject_to_pronoun_change = string[:-1]
        pronoun_changed_sentence = []
        for tokens in subject_to_pronoun_change:
            if tokens in ['나는', '나도', '내', '나의', '내가']:
                get_index = change_pronoun['Na_list'][change_pronoun['Na_list'] == tokens].index
                index = int(*get_index)
                tokens = change_pronoun['DangShin_list'][index]
            pronoun_changed_sentence.append(tokens)

        #eomi change
        word = string[-1]
        morphtags = Komoran().pos(word)

        if morphtags[0][1] == 'VA' or morphtags[0][1] == 'VV' or morphtags[0][1] == 'VXV' or morphtags[0][1] == 'VXA' or morphtags[0][1] == 'VCP' or morphtags[0][1] == 'VCN': #lemmatize
            changed_word = str(morphtags[0][0] + '군요.')
            changed_sentence = ' '.join(pronoun_changed_sentence) + ' ' + changed_word
            return changed_sentence

        else :
            candidate = ['그런 일이 있으셨군요.', '그랬군요.', '그러셨군요.']
            sentence = np.random.choice(candidate)

    return sentence
```
